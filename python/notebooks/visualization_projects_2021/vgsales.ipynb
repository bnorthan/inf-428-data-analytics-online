```python
import pandas as pd
```

```python
data=pd.read_csv('vgsales.csv')
```

```python
data=data.fillna(value=0)
```

```python
data.shape
```

```python
groups=data.groupby('Publisher')
```

```python
test=groups.Global_Sales.transform(len)
test
```

```python
# get the count of all the publishers and sort descending
count=groups.Name.count().sort_values(ascending=False)
count.head()
```

```python
# this is a bit confusing... but counting any column gives you the total number in each group...  try counting everything
# and note the 'count' of each column is the same... the mean, min, max etc of every column won't be the same, so the
# count of any column is the group size.
groups.count()
```

```python
# we can quickly generate a plot from the count by chaing head (take the first 10), plot (access the polling lib), and bar 
# (make a bar plot)
#matplotlibinline
groups.Name.count().sort_values(ascending=False).head(10).plot.bar(figsize=(19,9))
```
